Task 2 — Clean the Data

```python
df.duplicated().sum()
```

```python
df[df.duplicated(keep=False)].sort_values('car_name')
```

```python
df=df.dropna(subset=['selling_price'])
df.shape
```

```python
df['brand'] = df['brand'].str.strip().str.lower()
df['model'] = df['model'].str.strip().str.lower()
df['car_name'] = df['car_name'].str.strip().str.lower()
    
    
```

```python
df['mileage'] = df['mileage'].astype(str).str.extract('(\\d+\\.?\\d*)').astype(float)
df.head()
```

```python
df['engine'] = df['engine'].astype(str).str.extract('(\\d+)').astype(float)
```

```python
df['max_power'] = df['max_power'].astype(str).str.extract('(\\d+\\.?\\d*)').astype(float)
```

```python
num_cols = ['mileage', 'engine', 'max_power', 'seats']

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df['seats'] = df['seats'].fillna(df['seats'].median())
df['seats'] = df['seats'].astype(int)
```

```python
df['mileage'] = df['mileage'].fillna(df['mileage'].median())
df['engine'] = df['engine'].fillna(df['engine'].median())
df['max_power'] = df['max_power'].fillna(df['max_power'].median())
```

```python
print(df.isnull().sum())
print(df.info())
```

```python
df = df.drop_duplicates()
df.shape
```

```python
import numpy as np

# Step 1: Calculate mean selling price
mean_price = df['selling_price'].mean()

# Step 2: Create predictions (same value for all rows)
predictions = np.full(len(df), mean_price)

# Step 3: Calculate MAE
mae = np.mean(np.abs(df['selling_price'] - predictions))

print("Baseline Mean Price:", mean_price)
print("Baseline MAE:", mae)
```
